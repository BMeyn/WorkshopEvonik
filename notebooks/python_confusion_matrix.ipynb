```python
import pandas as pd
import seaborn as sns

data = {'y_Actual':    [1, 0, 0, 1, 0, 1, 0, 2, 1, 0, 1, 0, 2, 2],
        'y_Predicted': [1, 1, 0, 1, 0, 1, 1, 2, 1, 0, 0, 0, 1, 2]
        }
df = pd.DataFrame(data, columns=['y_Actual','y_Predicted'])
confusion_matrix = pd.crosstab(df['y_Actual'], df['y_Predicted'], rownames=['Actual'], colnames=['Predicted'])

sns.heatmap(confusion_matrix, annot=True)
```

```python
from sklearn.metrics import roc_curve, auc
import pandas as pd

import seaborn as sns

fpr, tpr, _ = roc_curve([1, 0, 1, 0, 1, 0, 0], [0.9, 0.8, 0.7, 0.7, 0.6, 0.5, 0.4])
roc_auc = auc(fpr, tpr)

df = pd.DataFrame({"True Positive Rate": tpr, "False Positive Rate": fpr})

ax = sns.lineplot(data=df, x="False Positive Rate", y = "True Positive Rate")
sns.lineplot([0, 1], [0, 1])
```
